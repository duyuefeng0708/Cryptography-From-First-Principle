## Creating Polynomial Rings

```sage
# Build a polynomial ring over the rationals
R.<x> = PolynomialRing(QQ)
f = x^3 - 2*x + 1
print(f, 'degree:', f.degree())
```

## Polynomial Arithmetic

```sage
# Add, multiply, and divide polynomials
R.<x> = PolynomialRing(QQ)
f = x^2 + 1
g = x - 1
print('f*g =', f*g)
print('divmod:', f.quo_rem(g))
```

## Evaluation and Roots

```sage
# Evaluate a polynomial and find its roots
R.<x> = PolynomialRing(QQ)
f = x^2 - 5*x + 6
print('f(2) =', f(2))
print('roots:', f.roots())
```
