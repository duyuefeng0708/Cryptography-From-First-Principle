## Ring Axioms

```sage
# Verify ring axioms for Z: additive group + multiplicative monoid
# Distributive: a*(b+c) == a*b + a*c
a, b, c = 3, 5, 7
assert a*(b+c) == a*b + a*c
```

## Z as a Ring

```sage
# SageMath's ZZ is the ring of integers
R = ZZ
print(R)
print(R.is_commutative())
```

## Rings vs Groups

```sage
# A ring has TWO operations; a group has one
R = Zmod(12)
print('Zero:', R.zero())
print('One:', R.one())
```
