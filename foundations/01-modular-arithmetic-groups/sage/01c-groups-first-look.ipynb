## Closure and the Binary Operation

```sage
# Verify closure: adding any two elements of Z_7 stays in Z_7
R = Zmod(7)
a, b = R(3), R(5)
print(a + b, a + b in R)
```

## Identity and Inverses

```sage
# Find the additive identity and inverses in Z_7
R = Zmod(7)
for x in R:
    print(f'inverse of {x} is {-x}')
```

## Checking the Group Axioms

```sage
# Verify associativity for a sample triple
R = Zmod(7)
a, b, c = R(2), R(4), R(6)
assert (a + b) + c == a + (b + c)
```
