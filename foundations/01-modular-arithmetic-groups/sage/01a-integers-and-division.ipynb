## The Division Algorithm

```sage
# Explore the division algorithm: a = q*b + r
a, b = 37, 7
q, r = divmod(a, b)
print(f'{a} = {q}*{b} + {r}')
```

## Remainders and Congruence

```sage
# Compute remainders for various values and observe patterns
for k in range(20):
    print(f'{k} mod 7 = {k % 7}')
```

## Computing with SageMath

```sage
# Use SageMath's Mod and divmod for exact arithmetic
print(divmod(100, 13))
print(Mod(100, 13))
```
